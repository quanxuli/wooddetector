```python
from ultralytics import YOLO

# Khởi tạo mô hình (ví dụ: YOLO26l Large, bạn có thể đổi thành model khác tuỳ ý)
model = YOLO('yolo26m.pt')

results = model.train(
    data="/content/datasets/data.yaml",  # Đã sửa lại đúng đường dẫn file YAML
    epochs=200,
    imgsz=1024,          # Độ phân giải cực cao cho bản M
    batch=32,            # Batch size lý tưởng cho A100 80GB
    optimizer='AdamW',
    lr0=0.001,
    cos_lr=True,
    patience=50,
    close_mosaic=20,     # Tắt Mosaic cuối để tinh chỉnh Bbox
    cache='disk',         # Nạp toàn bộ dữ liệu lên RAM để max tốc độ
    workers=12,          # Tận dụng tối đa số nhân CPU để nạp ảnh
    augment=True,
    overlap_mask=True,   # Nếu bạn làm segmentation
    val=True,            # Luôn kiểm tra trên tập Valid sau mỗi epoch
    project="/content/drive/MyDrive/runs",
    name="YOLOv26m",
    # Tăng cường dữ liệu mức độ cao
    mosaic=1.0,
    mixup=0.3,
    copy_paste=0.3
)
```
